### 7) Which was the most-ordered item? 

```python
chipo.item_name = chipo.item_name.str.replace('-', ' ')
quantities = {}
for item in chipo.item_name.unique():
    quantities[item] = chipo[chipo.item_name == item].quantity.sum()
items = list(quantities.keys())
quantities = list(quantities.values())
items[quantities.index(np.max(quantities))]
```

### 8) For the most-ordered item, how many items were ordered?

```python
np.max(quantities)
```

### 9) What was the most ordered item in the choice_description column?

```python
chipo.choice_description.describe().top
```

### 10) How many items were ordered in total?

```python
np.sum(quantities)
```

### 11) Turn the item price into a float

```python
chipo.item_price = chipo.item_price.str.replace('$', '').astype(float)
chipo.head()
```

### 12) How much was the revenue for the period in the dataset?
#### Revenue will be the sum of quantity * item_price!

```python
revenue = chipo.quantity * chipo.item_price
revenue.sum()
```

### 13) How many orders were made in the period?

```python
chipo.order_id.nunique()
```

### 14) What is the average revenue amount per order?

```python
revenue.sum() / chipo.order_id.nunique()
```
