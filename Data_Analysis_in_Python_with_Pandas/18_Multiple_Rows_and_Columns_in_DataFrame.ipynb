```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head(3)
```

```python
ufo.loc[0, :]
```

```python
ufo.loc[[0, 1, 2], :]
```

```python
ufo.loc[0:2, :] # loc is inclusive of right one
```

```python
ufo.loc[0:2]
```

```python
ufo.loc[:, ['City', 'State']]
```

```python
ufo.loc[:, 'City':'State']
```

```python
ufo.loc[0:2, 'City':'State']
```

```python
ufo.head(3).drop('Time', axis = 1)
```

```python
ufo[ufo.City == 'Oakland']
```

```python
ufo.loc[ufo.City == 'Oakland', :]
```

```python
ufo.loc[ufo.City == 'Oakland', 'State'] # takes only one operation
```

```python
ufo.loc[ufo.City == 'Oakland'].State # Takes more than one operation
```

```python
ufo.iloc[:, [0, 3]]
```

```python
ufo.iloc[:, 0:4] # iloc is exclusive of right hand side number
```

```python
ufo.iloc[0:3, :]
```

```python
ufo[['City', 'State']]
```

```python
ufo.loc[:, ['City', 'State']]
```

```python
ufo[0:2]
```

```python
ufo.iloc[0:2, :]
```
