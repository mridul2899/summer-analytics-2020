```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head()
```

```python
ufo.shape
```

```python
# drop method
ufo.drop('Colors Reported', axis = 1, inplace = True)
# axis = 1 specifies that a column has to be removed
# inplace = True specifies that the changes have to be inplace,
# i.e., the changes must be made into the same dataframe
```

```python
ufo.head()
```

```python
# drop multiple methods
# pass a list of strings containing column names
ufo.drop(['City', 'State'], axis = 1, inplace = True)
```

```python
# For dropping rows,
ufo.drop([0, 1], axis = 0, inplace = True)
# list of labels/indices is passed as first argument
# the list contains all the rows that are to be dropped
# axis = 0 specifies that rows are to be dropped
# by default, axis is equal to 0, so we need not specify
# axis = 0 explicitly
```
