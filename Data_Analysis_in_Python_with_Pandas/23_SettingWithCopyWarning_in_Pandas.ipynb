```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies.content_rating.isnull().sum()
```

```python
movies[movies.content_rating.isnull()]
```

```python
movies.content_rating.value_counts()
```

```python
movies[movies.content_rating == 'NOT RATED']
```

```python
movies[movies.content_rating == 'NOT RATED'].content_rating
```

```python
import numpy as np
```

```python
movies[movies.content_rating == 'NOT RATED'].content_rating = np.nan
```

```python
movies.content_rating.isnull().sum()
```

```python
# no warning with loc
movies.loc[movies.content_rating == 'NOT RATED', 'content_rating'] = np.nan
```

```python
movies.content_rating.isnull().sum()
```

This is because in the previous method, there is a get item part and a set item part.<br>
Pandas cannot guarantee whether the first part returned a view of the data or a copy of the data. <br>
If view of the data: change will be represented in the data. <br>
If copy: change will be made, but in some copy and not in the actual data. <br>
Since Pandas cannot assure which one it is, it shows the warning. <br>
The Pandas loc rather makes this whole operation as a single set item operation.

```python
top_movies = movies.loc[movies.star_rating >= 9, :]
```

```python
top_movies
```

```python
top_movies.loc[0, 'duration'] = 150
```

```python
top_movies # it worked despite the warning
```

```python
# warning was released because it wasn't sure between view and copy for top_movies
# Solution: explicitly state while creating a copy
top_movies = movies.loc[movies.star_rating >= 9, :].copy()
```

```python
top_movies.loc[0, 'duration'] = 150
```

```python
top_movies
```

```python
movies.head() # no change made in movies dataframe
```
