```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies['title'].sort_values()
```

```python
movies['title'].sort_values(ascending = False)
```

The above methods do not sort the dataframe.

```python
movies.sort_values('title') # does not affect the original dataframe
```

```python
movies.sort_values('duration')
```

```python
movies.sort_values('duration', ascending = False)
```

```python
movies.head() # We can see that the above methods do not change the dataframe
```

```python
# Sorts the dataframe by content rating followed by duration,
# i.e., ties in content ratings are broken by duration values
movies.sort_values(['content_rating', 'duration'])
```
