```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head()
```

```python
ufo.columns # gives column names
```

```python
# rename method, dictionary passed to columns
# keys are the old column names
# values are the new column names
# multiple columns can be renamed at once
# inplace = True used for making the changed to ufo dataframe
ufo.rename(columns = {'Colors Reported': 'Colors_Reported', 'Shape Reported': 'Shape_Reported'}, inplace = True)
```

```python
ufo.columns
```

```python
ufo_cols = ['city', 'colors reported', 'shape reported', 'state', 'time']
ufo.columns = ufo_cols
# helps in renaming all the column names, as specified in order by the list
# hence reduces the need for typing old column names
```

```python
ufo.head()
```

```python
# Or, you may rename the column names while reading the CSV file itself, by changing the names attribute
# header = 0 is also passed to specify that 0th row already contains columns names in the csv file,
# which however are not to be used here, and instead the new ones are to be used
ufo = pd.read_csv('http://bit.ly/uforeports', names = ufo_cols, header = 0)
```

```python
ufo.head()
```

```python
ufo.columns
```

```python
ufo.columns = ufo.columns.str.replace(' ', '_') # string method
# replaces spaces from all the column names with underscores
```
