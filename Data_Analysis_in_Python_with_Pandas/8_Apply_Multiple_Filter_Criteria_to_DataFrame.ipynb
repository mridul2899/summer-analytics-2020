```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies[movies.duration >= 200]
```

```python
# For multiple filter criteria, we use logical operations
# and, or, not
# we cannot use and here, we rather use &
movies[(movies.duration >= 200) & (movies.genre == 'Drama')]
```

```python
# or condition - pipe character '|' instead of or word
movies[(movies.duration >= 200) | (movies.genre == 'Drama')]
```

```python
(movies.duration >= 200) & (movies.genre == 'Drama')
```

```python
# For multiple acceptable genres,
movies[(movies.genre == 'Crime') | (movies.genre == 'Drama') | (movies.genre == 'Action')]
```

```python
# the above trick is very tedious
# better method:
movies[movies.genre.isin(['Crime', 'Drama', 'Action'])]
```
