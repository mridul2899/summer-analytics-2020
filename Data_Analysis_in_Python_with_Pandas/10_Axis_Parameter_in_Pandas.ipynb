# Axis Parameter in Pandas

```python
import pandas as pd
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
drinks.head()
```

```python
drinks.drop('continent', axis = 1).head()
```

```python
drinks.drop(2, axis = 0).head()
```

```python
drinks.mean() # axis = 0 by default
```

axis = 0 means that operation moves along the rows <br>
axis = 1 means that operation moves along the columns

```python
drinks.mean(axis = 0)
```

```python
drinks.mean(axis = 1) # calculating average across the rows
```

The above quantities are not meaningful because the quantities differ, units differ et cetera. <br>
This was only done to show the usage of axis parameter.

```python
drinks.drop(2, axis = 0).head()
```

```python
drinks.mean(axis = 0).shape
```

```python
drinks.mean(axis = 'index') # axis = 'index' is same as axis = 0
```

```python
drinks.mean(axis = 'columns') # axis = 'columns' is same as axis = 1
```
