'movies' is a dataframe, it has some methods and some attributes. <br>
Methods must have parentheses, while attributes do not need have parentheses. <br>
Methods can be considered as some kind of actions. <br>
Attributes can be considered as some kind of descriptions. <br>

```python
movies.describe(include = ['object']) # even some attributes may use parentheses
# for passing additional arguments
# Like in this example, only the series with datatype 'object' (like strings) will be described
```

Some useful **shortcuts**:
1. Use tab after a dataframe (object in general) to see all its attributes and methods (indistinguishable though).
2. Use Shift + Tab to see the description of any object, function, variable etc. Press multiple times for more detailed versions/longer duration versions.
