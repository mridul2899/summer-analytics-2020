```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports') # just like read table, except it is for comma separated values file
```

```python
type(ufo)
```

```python
ufo.head()
```

```python
ufo['City']
```

```python
type(ufo['City'])
```

```python
ufo.City # python makes column name (series name) an attribute of the dataframe
# this attribute makes it easier to access series in dataframes, because of lesser typing involved.
```

```python
# dot notation does not work for space separated values
# You cannot use ufo.Colors Reported or any other separator with dot convention
ufo['Colors Reported']
```

```python
# ufo.shape, if column name would have been shape instead of shape recorded,
# will not work as shape already is an attribute of dataframe
# same for ufo.head for some column name head in a dataframe
# hence for all these cases, it is better to use the square bracket notation
ufo.shape
```

```python
ufo.City + ufo.State # it just concatenates city names with state names
```

```python
# better way would be to write it like:
ufo.City + ', ' + ufo.State
```

```python
# Now, to make an additional series in this dataframe,
# although we will have to use square bracket notation
# for the name of the new series, since its attribute
# hasn't already been made in the dataframe by Python
ufo['Location'] = ufo.City + ', ' + ufo.State
```
